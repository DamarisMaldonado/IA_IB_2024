```python
#Importamos librerias
import cv2
import numpy as np
import matplotlib.pyplot as plt


# cargamos las dos imagenes en escala de grises
imagen2 = cv2.imread("/content/mano1.jpeg", cv2.IMREAD_GRAYSCALE)
imagen1 = cv2.imread("/content/mano2.jpeg", cv2.IMREAD_GRAYSCALE)

imagen2_res = cv2.resize(imagen2, (imagen1.shape[1], imagen1.shape[0]))

# Aplicar la suma ponderada con pesos =0.7 y =0.3
alpha = 0.3  # Peso de la primera imagen
beta = 0.7   # Peso de la segunda imagen
gamma = 10     # Ajuste de brillo

imagen_resultado = cv2.addWeighted(imagen1, alpha, imagen2_res, beta, gamma)
# Mostrar imgenes
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(imagen1, cmap="gray")
plt.title("Imagen Original 1")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(imagen2, cmap="gray")
plt.title("Imagen Original 2")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(imagen_resultado, cmap="gray")
plt.title("Resultado Suma Ponderada")
plt.axis("off")
plt.show()



```

```python
imagenes = ["/content/mano1.jpeg", "/content/mano2.jpeg"]
imagenes_cargadas = [cv2.imread(img, cv2.IMREAD_GRAYSCALE) for img in imagenes]

# Verificar que todas las imgenes tengan el mismo tamao
alto, ancho = imagenes_cargadas[0].shape
imagenes_cargadas = [cv2.resize(img, (ancho, alto)) for img in imagenes_cargadas]

# Crear una imagen acumuladora en formato flotante
suma = np.zeros((alto, ancho), dtype=np.float32)

# Sumar todas las imgenes
for img in imagenes_cargadas:
    suma += img

# Calcular el promedio dividiendo por la cantidad de imgenes
imagen_promedio = suma / len(imagenes_cargadas)

# Convertir la imagen a formato uint8 para visualizacin
imagen_promedio = cv2.convertScaleAbs(imagen_promedio)

# Mostrar imgenes originales
for i, img in enumerate(imagenes_cargadas):
    plt.subplot(1, len(imagenes), i+1)  # Crear subgrficos en una fila
    #plt.figure(figsize=(10, 5))
    plt.imshow(img, cmap="gray")  # Mostrar imagen
    plt.title(f"Imagen {i+1}")  # Ttulo de cada imagen
    plt.axis("off")  # Ocultar ejes
plt.show()
#Mostrar imagen promedio
plt.figure(figsize=(5, 5))
#plt.subplot(1, 3, 1)
plt.imshow(imagen_promedio, cmap="gray")
plt.title("Imagen Promedio")
plt.axis("off")
plt.show()
```

```python
# Cargar las dos imgenes
img1 = cv2.imread("/content/mano1.jpeg", cv2.IMREAD_GRAYSCALE)  # Cargar en escala de grises
img2 = cv2.imread("/content/mano2.jpeg", cv2.IMREAD_GRAYSCALE)

# Asegurar que tengan el mismo tamao
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Realizar la resta de imgenes
resta = cv2.subtract(img2, img1)  # OpenCV maneja el rango automticamente

# Mostrar imgenes originales y resultado
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title("Imagen 1")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
plt.title("Imagen 2")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(resta, cmap='gray')
plt.title("Imagen Restada")
plt.axis("off")

plt.show()
```

```python
# Cargar imagen en escala de grises
img = cv2.imread("/content/mano1.jpeg", cv2.IMREAD_GRAYSCALE)

# Definir el factor de multiplicacin
factor = 1.5  # Aumenta el brillo en un 50%

# Multiplicar la imagen por el factor
img_multiplicada = cv2.multiply(img, np.array([factor], dtype=np.float32))

# Convertir la imagen a uint8 para visualizacin correcta
img_multiplicada = cv2.convertScaleAbs(img_multiplicada)

# Mostrar la imagen original y la modificada
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Imagen Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img_multiplicada, cmap='gray')
plt.title(f"Imagen x {factor}")
plt.axis("off")

plt.show()
```

```python
# Cargar las dos imgenes en escala de grises
img1 = cv2.imread("/content/mano1.jpeg", cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("/content/mano2.jpeg", cv2.IMREAD_GRAYSCALE)

# Redimensionar para que tengan el mismo tamao
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Multiplicacin elemento a elemento
img_multiplicada = cv2.multiply(img2, img1)  # OpenCV maneja el rango automticamente

# Convertir la imagen resultante a uint8
img_multiplicada = cv2.convertScaleAbs(img_multiplicada)

# Mostrar imgenes originales y resultado
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title("Imagen 1")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')
plt.title("Imagen 2")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(img_multiplicada, cmap='gray')
plt.title("Multiplicacin Elemento a Elemento")
plt.axis("off")

plt.show()
```

```python
# Crear dos imgenes binarias (Ejemplo: cuadrados blancos sobre fondo negro)
img1 = np.zeros((200, 200), dtype=np.uint8)
img2 = np.zeros((200, 200), dtype=np.uint8)

cv2.rectangle(img1, (50, 50), (150, 150), 255, -1)  # Cuadrado blanco en img1
cv2.circle(img2, (100, 100), 50, 255, -1)  # Crculo blanco en img2

# Aplicar operaciones lgicas
and_img = cv2.bitwise_and(img1, img2)  # AND lgico
or_img = cv2.bitwise_or(img1, img2)    # OR lgico
not_img1 = cv2.bitwise_not(img1)       # NOT lgico en img1
xor_img = cv2.bitwise_xor(img1, img2)  # XOR lgico

# Mostrar las imgenes
titles = ["Imagen 1", "Imagen 2", "AND", "OR", "NOT (Imagen 1)", "XOR"]
images = [img1, img2, and_img, or_img, not_img1, xor_img]

plt.figure(figsize=(12, 6))

for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis("off")

plt.show()
```
